```python
#Dropping the outliers. 
df = df[(df["Area Population"]<60000)&(df["Area Population"]>10000)]
df = df[(df["Avg. Area Number of Rooms"]<10)&(df["Avg. Area Number of Rooms"]>4)]
df = df[(df["Avg. Area House Age"]<9)&(df["Avg. Area House Age"]>3)]
df = df[(df["Avg. Area Income"]<100000)&(df["Avg. Area Income"]>40000)]
df.shape
```

```python
df.columns
```

```python
X=df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.4,random_state=101)
```

```python
#model training
```

```python
from sklearn.linear_model import LinearRegression #import linear regression
```

```python
lm=LinearRegression() #create an object
```

```python
lm.fit(X_train,y_train) #fit the data
```

```python
lm.coef_
```

```python
coef=pd.DataFrame(lm.coef_,X.columns,columns=['Coeff'])
```

```python
coef
```

```python
predictions=lm.predict(X_test)
```

```python
plt.scatter(y_test,predictions)
```

```python
import cufflinks as cf 
cf.go_offline()
import plotly.io as pio
pd.DataFrame({"Giá dự đoán":predictions, "Giá ban đầu": y_test}).iplot(mode="markers")
```

```python
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, predictions))
print('MSE:', metrics.mean_squared_error(y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
metrics.explained_variance_score(y_test, predictions)
```
